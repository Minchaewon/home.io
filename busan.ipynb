```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumBarunGothic'
```

```python
# 데이터 불러오기
filename  = 'C:/Users/mcw08/OneDrive/바탕 화면/OUTLIER(23상반기)/travel/부산여행.csv'

df_busan = pd.read_csv(filename,  encoding='cp949')

df_busan
```

```python
import folium
from folium.plugins import MiniMap, MarkerCluster

# 부산 중심 좌표 설정
Busan_location = [35.1796, 129.0756]

# 지도 생성
map_Busan_travel = folium.Map(location=Busan_location, zoom_start=10)

# 미니맵 생성
minimap = MiniMap(tiles='OpenStreetMap', width=150, height=150, toggle_display=True)
map_Busan_travel.add_child(minimap)

# MarkerCluster 객체 생성
marker_cluster = MarkerCluster().add_to(map_Busan_travel)

# df_Busan의 각 행을 순회하며 장소 정보를 클러스터에 추가
for index, row in df_busan.iterrows():
    place_name = row['PLACE_NM']
    place_type = row['PLACE_TY']
    latitude = row['LC_LA']
    longitude = row['LC_LO']
    addr_list = row['ADDR']
    place_link = row['LINK']

    # 팝업 내용 구성
    popup_text = f'장소명: {place_name}<br>장소 유형: {place_type}<br><a href="{place_link}" target="_blank">장소 링크 바로가기</a><br>주소:'
    for addr in addr_list:
        popup_text += f'{addr}'

    # 마커 생성 및 클러스터에 추가
    marker_color = 'green' if place_type == 'playground' else 'orange' if place_type == 'hotel' else 'blue'
    folium.Marker(
        location=[latitude, longitude],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color=marker_color, icon='info-sign')
    ).add_to(marker_cluster)

# 지도 출력
map_Busan_travel
```

```python
# HTML 파일로 저장
map_Busan_travel.save("map_Busan_travel.html")

# 웹브라우저에서 HTML 파일 열기
import webbrowser
webbrowser.open("map_Busan_travel.html")
```
